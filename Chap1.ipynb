## 1.4 Localiser les extrema d'une fonction : exemple

Représentez le graphe de la fonction
\begin{equation}
f:[-3,3]\rightarrow\mathbb{R}:x\rightarrow(1+x)\exp(-x^2+3x\cos(x))-(1+x^4)^2\sin(x)
\end{equation}
Dessinez le graphe en rouge et en pointillés.<br /> 
Ajoutez une grille à votre graphe.<br />
Trouvez tous les extrema de la fonction sur $[-3,3]$ à l'aide de la fonction prédéfinie `fmin` de scipy.optimize<br />
Evaluez la fonction en ses extrema : donnez leurs coordonnées.<br />
Ajoutez au graphe une croix sur chaque extremum, et notez "minimum" ou "maximum" en dessous à l'aide de la fonction prédéfinie `text` de matplotlib.pyplot.  

```python
def f(x): return (1+x)*np.exp(-x**2+3*x*np.cos(x))-(1+x**4)**2*np.sin(x)
# Graphe
x=np.linspace(-3,3,1000)
plt.plot(x,f(x),"r:")
plt.grid()
# Minimum de la fonction
xmin=sciop.fmin(f,3)   # 3 = first guess (valeur approchée de l'abscisse du minimum, par simple inspection visuelle du graphe). Cette fonction renvoie avec grande précision le x tel que f(x) est min.
print("Le minimum de la fonction a pour coordonnées (",xmin,",",f(xmin),")")
plt.plot(xmin,f(xmin),"kx",markersize=10)
plt.text(xmin-1,f(xmin),"Minimum",color="m")
# Maximum de la fonction
def g(x): return -f(x)
xmax=sciop.fmin(g,-3)      # -3 = first guess. Renvoie avec grande précision le x tel que -f(x) est min, c-à-d tel que f(x) est max 
# les 2 lignes précédentes peuvent s'écrire + simplement en 1 ligne: xmax=sciop.fmin(lambda x: -f(x),-3)
print("Le maximum de la fonction a pour coordonnées (",xmax,",",f(xmax),")")
plt.plot(xmax,f(xmax),"kx",markersize=10)
plt.text(xmax+0.2,f(xmax),"Maximum",color="m")
```

## 1.5 Ajouter une légende et localiser les racines : exemple

Représentez dans un même diagramme, les graphes des fonctions 
\begin{equation}
f(x)=\sin(x^2)\quad\textrm{et}\quad g(x)=\frac{\exp(x)}{4}
\end{equation}
pour $x\in[0,2]$.<br />
Ajoutez une légende.<br />
Localisez grossièrement à l'aide de la figure les valeurs de $x$ telles que $\sin(x^2)=\frac{\exp(x)}{4}$. Calculez-les ensuite précisément avec la fonction prédéfinie `scipy.optimize.fsolve`. 

```python
def f(x): return np.sin(x**2)
def g(x): return np.exp(x)/4
x=np.linspace(0,2,1000)
plt.plot(x,f(x),label="sin(x**2)")
plt.plot(x,g(x),label="exp(x)/4")
plt.legend()
plt.grid()
```

f(x)=g(x) lorsque h(x)=f(x)-g(x)=0. Il suffit donc de chercher les racines de la fonction h, à l'aide de la fonction prédéfinie `fsolve` de scipy optimize. Comme pour `fmin`, il faut donner à cette fonction un first guess après inspection visuelle sur le graphe. 

```python
def h(x): return f(x)-g(x)
racine1=sciop.fsolve(h,0.75)
racine2=sciop.fsolve(h,1.3)
print("Entre 0 et 2, les deux racines sont ",racine1," et ",racine2)
```
